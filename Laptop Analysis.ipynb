# Total Models by Brands

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x='brand', data=df, palette='magma')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')    
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Total Models By Brands')  
plt.xlabel('Brands')
plt.ylabel('Total Models')
sns.set_style("whitegrid")
plt.grid(False)
plt.show()    
```

# Total Models by Primary Storage Capacity

```python
model_by_prim_sto = df.groupby('primary_storage_capacity')['Model'].count().sort_values(ascending=False)
model_by_prim_sto 
```

```python
sns.barplot(y=model_by_prim_sto.values, x=model_by_prim_sto.index, orient='vertical', palette='rocket')
plt.title('Total Models By Primary Storage Capacity')  
plt.xlabel('Primary Storage Capacity')
plt.ylabel('Total Models')
sns.set_style("whitegrid")
plt.grid(False)
plt.show()
```

# Total Brands By Average Price

```python
model_by_avg_price = df.groupby(['brand'])['Price'].mean().sort_values(ascending=False)
model_by_avg_price 
```

```python
model_by_avg_price_rounded = model_by_avg_price.round(2)
```

```python
plt.figure(figsize=(10,10))
barplot7=sns.barplot(x=model_by_avg_price_rounded.values, y=model_by_avg_price_rounded.index, orient='horizontal', palette='plasma')
plt.bar_label(barplot7.containers[0], labels=model_by_avg_price_rounded.values, label_type='edge')
plt.title('Total Brands By Average Price')  
plt.ylabel('Average Price')
plt.xlabel('Total Brands')
plt.xlim(0,220000)
sns.set_style("whitegrid")
plt.grid(False)
plt.show()
```

# Insights

ASUS is offering highest number of models

 Windows is the common Operating System
 
 Most Process Brand is by Intel and Process Tier is mostly by core i5, followed by core i7
 
 Primary storage is SSD, whereas secondary storage is not present in 98% of the laptops
 
 Primary storage capacity 512 is available in most laptops 
 
 Touch Screen is not offered in most of of laptops, only in 8.8% laptops
 
 Laptop which is most expensive is from DELL exclusively made for gaming purpose and if talk about average  pricing Apple laptops are most expensive
 
 Average warranty of laptops are 1 year including most expensive laptop and mac laptops, whereas Lenevo is  providing warranty for 3 years, mostly having windows OS
 
 Ram memory of most the laptops is 16 GB, followed by 8 GB
 
 Common Resolution(height) is 1080 ,  Resolution(width) is 1920
 
 GPU brand mostly by Intel followed by nvidia
 
 Dell Alienware m16 Gaming Laptop (AMD Ryzen 9 7845HX/ 32GB/ 2TB SSD/ Win 11/ 12GB Graph) might be the best laptop as it has window OS, 32 GB Ram with primary storage capacity of 2048 but its name suggests it is best suited for gaming 
 
 For budget friendly, normal tasks with primary capacity of 1024, 16 GB, window OS & rating 62 is Acer Aspire Lite AL15-41 Laptop (AMD Ryzen 5 5500U/ 16GB/ 1TB SSD/ Win11)
